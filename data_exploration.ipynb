## Data Wrangling
### Parse Tabular Table

````python
import re
import pandas as pd
from io import StringIO

# Function to parse the question text and extract the tables
def parse_question_tables(question_text):
    # Extract user plane table
    user_match = re.search(
        r"User plane drive test data as follows：\n\n(.*?)\n\n\nEngeneering parameters data as follows：",
        question_text,
        re.S
    )

    # Extract engineering table
    eng_match = re.search(
        r"Engeneering parameters data as follows：\n\n(.*)",
        question_text,
        re.S
    )

    user_table_text = user_match.group(1).strip()
    eng_table_text = eng_match.group(1).strip()

    user_df = pd.read_csv(StringIO(user_table_text), sep="|")
    eng_df = pd.read_csv(StringIO(eng_table_text), sep="|")

    return user_df, eng_df
````

````python
user_df, eng_df = parse_question_tables(df.loc[0, "question"])

user_df.head()
````

# Telecommunication User Plane Table – Column Definitions

## Summary of Telecommunication User Plane Columns

| Column | What it means | Simple intuition | Likely importance |
|----------|----------------|------------------|-------------------|
| `Timestamp` | Time of measurement | When the signal was recorded | Useful for time patterns |
| `Longitude`, `Latitude` | GPS coordinates | Where the car/device was | Important for mapping weak areas |
| `GPS Speed (km/h)` | Vehicle speed | How fast the user is moving | High speed can hurt signal stability |
| `5G KPI PCell RF Serving PCI` | Serving cell/tower ID | Which 5G tower the phone is connected to | Very important |
| `5G KPI PCell RF Serving SS-RSRP [dBm]` | Signal strength | *"How strong is the signal?"* | Very important |
| `5G KPI PCell RF Serving SS-SINR [dB]` | Signal quality/interference | *"How clean is the signal?"* | Extremely important |
| `5G KPI PCell Layer2 MAC DL Throughput [Mbps]` | Download throughput | Actual internet speed | Likely target variable |
| `Measurement PCell Neighbor Cell Top Set(Cell Level) Top 1–5 PCI` | Neighboring towers | Nearby cells competing/helping | Useful for interference and handover analysis |
| `Measurement PCell Neighbor Cell Top Set(Cell Level) Top 1–5 Filtered Tx BRSRP [dBm]` | Neighbor signal strengths | Strength of nearby towers | Important |
| `5G KPI PCell Layer1 DL RB Num (Including 0)` | Resource blocks allocated | How much radio bandwidth the user received | Extremely important |

### Quick Intuition

- **RSRP** → Signal strength  
- **SINR** → Signal quality / interference  
- **PCI** → Tower/cell identifier  
- **Neighbor PCI/BRSRP** → Nearby competing towers  
- **RB Num** → Amount of network resources assigned  
- **Throughput** → Actual internet speed experienced by the user  
- **GPS Speed** → Mobility effect on network stability

Since this dataset comes from **5G wireless network drive-test logs**, many variables represent **signal strength, signal quality, neighboring cell behavior, mobility, and network resource allocation**.  

The goal appears to be understanding **why throughput drops below 600 Mbps**, so understanding these KPIs is important before modeling.

## 1. Timestamp

### `Timestamp`

**Definition:**  
The time at which the network measurement was recorded.

**Business Meaning:**  
Represents when the device measured network performance during the drive test.

**Why it matters:**  
Can help identify temporal patterns or sequence-based behaviors (e.g., throughput drops after handovers).

---

## 2. Location Information

### `Longitude`  
### `Latitude`

**Definition:**  
GPS coordinates of the test vehicle.

**Business Meaning:**  
Represents the physical location where the measurement was taken.

**Why it matters:**  
Useful for identifying:

- Geographic weak coverage zones
- Areas with interference
- Specific road sections with poor throughput

Example use case:
> Throughput consistently drops near a certain location.

---

## 3. Vehicle Speed

### `GPS Speed (km/h)`

**Definition:**  
The speed of the test vehicle during the measurement.

**Business Meaning:**  
Represents how fast the user is moving through the network.

**Why it matters:**  
Higher speeds may negatively impact network performance due to:

- Frequent cell switching (handovers)
- Reduced signal stability

Typical interpretation:

| Speed | Impact |
|--------|--------|
| Low speed | Stable connection |
| High speed (>40 km/h) | Possible throughput degradation |

Related root cause:
- **C7:** Test vehicle speed exceeds 40 km/h.

## 4. Serving Cell Identifier

### `5G KPI PCell RF Serving PCI`

**Definition:**  
PCI (**Physical Cell Identifier**) of the serving cell.

**Business Meaning:**  
Represents which 5G cell tower (or sector) the device is currently connected to.

**Why it matters:**  

Changes in PCI indicate **handover events**, where the device switches towers.

Frequent PCI switching may indicate:

- Poor network stability
- Excessive handovers

Related root cause:
- **C5:** Frequent handovers degrade performance.

Example:

```text
712 → connected to tower A
258 → connected to tower B
71 → connected to tower C
```

---

## 5. Signal Strength

### `5G KPI PCell RF Serving SS-RSRP [dBm]`

**Definition:**  
**SS-RSRP (Synchronization Signal Reference Signal Received Power)** measures signal strength from the serving cell.

**Business Meaning:**  
Represents how strong the network signal is.

**Why it matters:**  
Weak signal strength often leads to poor throughput.

Typical interpretation:

| RSRP (dBm) | Signal Quality |
|------------|----------------|
| > -80 | Excellent |
| -80 to -90 | Good |
| -90 to -100 | Fair |
| < -100 | Weak |

Examples:

```text
-77 dBm → Strong signal
-88 dBm → Moderate signal
-105 dBm → Weak signal
```

Potential causes of weak signal:

- **C1:** Excessive downtilt angle
- **C2:** Coverage overshooting

---

## 6. Signal Quality / Interference

### `5G KPI PCell RF Serving SS-SINR [dB]`

**Definition:**  
**SS-SINR (Signal-to-Interference-plus-Noise Ratio)** measures signal quality.

**Business Meaning:**  
Represents how clean the signal is relative to surrounding interference.

**Why it matters:**  

A signal can be strong but still perform poorly if interference is high.

Typical interpretation:

| SINR (dB) | Quality |
|------------|----------|
| > 20 | Excellent |
| 13–20 | Good |
| 0–13 | Moderate |
| < 0 | Poor |

Examples:

```text
19.8 dB → Excellent signal quality
6.6 dB → Moderate quality
1.81 dB → High interference likely
```

Potential root causes:

- **C4:** Overlapping coverage interference
- **C6:** PCI interference

**Why it matters:**  
Multiple cells can belong to the same tower.

Unique identifier for a cell sector within a base station.

A single tower usually serves multiple directions.

Think:

## 5. Mechanical Downtilt

Physical downward tilt of the antenna.

Controls how much the signal points downward.

One of the most important telecom variables.

Too little downtilt:

- **C1:** Downtilt too large


````python
parsed_rows = []

for i, row in df.iterrows():
    user_df, eng_df = parse_question_tables(row["question"])

    user_df["example_id"] = i
    eng_df["example_id"] = i
    user_df["answer"] = row["answer"]
    eng_df["answer"] = row["answer"]

    parsed_rows.append((user_df, eng_df))

all_user_data = pd.concat([x[0] for x in parsed_rows], ignore_index=True)
all_engineering_data = pd.concat([x[1] for x in parsed_rows], ignore_index=True)
````
